# Soal Nomor 1 dan 2

```python
import pandas as pd
from sklearn.model_selection import train_test_split

data = pd.read_csv('MovieData-2018-2020.csv')

population_df = pd.DataFrame(data['genre'])

genre_counts = population_df['genre'].value_counts() # kode ini digunakan untuk menghitung frekuensi tiap masing-masing genre muncul

valid_genres = genre_counts[genre_counts >= 2].index # baris ini digunakan untuk menyaring perhitungan dari kode diatas. jadi baris yang kurang dari 2 akan langsung dicoret

filtered_df = population_df[population_df['genre'].isin(valid_genres)] # baris ini bertujuan untuk menyaring dataset agar sesuai dengan ketentuan dan mengeceknya kembali

# Lakukan stratified sampling pada data yang sudah difilter
_, stratified_sample = train_test_split(
    filtered_df,
    test_size=0.2,
    stratify=filtered_df['genre'],
    random_state=59
)

print('Stratified Sampling:\n', stratified_sample.head(30)) # menampilkan 30 sampel
```

```python
import scipy.stats as stats
import math

x_bar = 5.304560e+07    #Rata-rata
s = 6.243908e+07       #STD Sample
n = 274          #Jumlah Data
confidence_level = 0.95 #Pilih antara 0.90/0.95/0.99

# calculate alpha, degrees of freedom(df), the critical t-value
alpha = (1-confidence_level)
df = n - 1
standard_error = s/math.sqrt(n)
critical_t = stats.t.ppf(1 - alpha/2, df)
margin_of_error = critical_t * standard_error

# Calculate the lower and upper bound of the confidence interval
lower_bound = x_bar - margin_of_error
upper_bound = x_bar + margin_of_error

# Print the results
print("Critical t-value: {:.2f}".format(critical_t))
print("Margin of Error: {:.2f}".format(margin_of_error))
print("Confidence Interval: [{:.2f},{:.2f}]".format(lower_bound,upper_bound))
print("The {:.1%} confidence interval for the population mean is:".format(confidence_level))
print("between {:.2f} and {:.2f}".format(lower_bound,upper_bound))
```
